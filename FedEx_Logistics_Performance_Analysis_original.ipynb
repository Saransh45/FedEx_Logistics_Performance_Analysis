## 3. ***Data Wrangling***

### Data Wrangling Code

````python
df1 = df.copy()
df1.isnull().sum()
````

````python
categorical_cols = df1.select_dtypes(include='object').columns
df1[categorical_cols] = df1[categorical_cols].fillna('Unknown')
numerical_cols = df1.select_dtypes(include=['int64','float64']).columns
df1[numerical_cols] = df1[numerical_cols].fillna(df1[numerical_cols].median())
````

````python
df1['Delivery Days'] = (
    pd.to_datetime(df1['Delivered to Client Date']) -
    pd.to_datetime(df1['Scheduled Delivery Date'])
).dt.days
df1['Year'] = pd.to_datetime(df1['Scheduled Delivery Date']).dt.year
df1['Month'] = pd.to_datetime(df1['Scheduled Delivery Date']).dt.month_name()
````

### What all manipulations have you done and insights you found?

Created a copy of the original dataset to preserve raw data.
Checked the dataset for missing values and data quality issues.
Replaced missing values in categorical columns with "Unknown" to avoid loss of records.
Replaced missing values in numerical columns using the median to minimize the impact of outliers.
Converted date columns into datetime format for time-based analysis.
Created a new feature Delivery Days by calculating the difference between scheduled delivery date and actual delivery date.
Extracted Year and Month from the scheduled delivery date to support trend analysis.

Insights Found During Data Wrangling:

The dataset contains both categorical and numerical variables related to shipment operations, product details, and logistics costs.
Missing values were present in several columns and required preprocessing before analysis.
Date-related fields provided an opportunity to analyze delivery performance and shipment trends over time.
Freight cost, shipment weight, and line item value were identified as important numerical variables for further analysis.
Shipment mode, country, vendor, and product group were identified as key categorical variables influencing logistics performance
